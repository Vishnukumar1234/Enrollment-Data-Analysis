```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Cleaned_data.xlsx')
df
```

```python
#Grouping the dataframe by Term and then aggregating the grade and Student column by their average and count

#Term wise student enrollments and their consolidated grade average

gdf = df.groupby('Term').agg({'Grade': 'mean','Student':pd.Series.count})
gdf
```

```python
#Grouping by the Unique Student Id followed by the term and then aggregating the Number of courses he has taken and his grade 
sdf = df.groupby(['Student','Term']).agg({'Course Prefix':pd.Series.count,'Grade':'mean'}).reset_index()
sdf
```

```python
#Validating the sdf on whether it equals the total row count
sum(sdf['Course Prefix'])
```

```python
#Using sdf to group by based on number of course taken and then aggregating their grade average
cdf = sdf.groupby(['Course Prefix','Term']).agg({'Grade':'mean'})
cdf
```

```python
cdf.to_csv("course_count_vs_grade.csv")
```

```python
#Grouping by Course Prefix and Student id columns which only has the value 'Y' in Credit Converted Column
subyy = df[df['Credit Converted']=='Y'].groupby(['Course Prefix','Student']).agg({'Grade':'mean'})
subyy
```

```python
subyy.to_csv("subjwise_grade.csv")
```

```python
#Converting the blank value in Credit Converted Column to N for further analysis

df['Credit Converted'].fillna('N',inplace=True)
df
```

```python
#Creating a new dataframe which has only the Value 'Y' in Credit Eligible column and then grouping by using Credit Converted, Course Prefix Column
conv = df[df['Credit Eligible']=='Y'].groupby(['Credit Eligible','Credit Converted','Course Prefix']).agg({'Student':pd.Series.count}).reset_index()
```

```python
#Splitting the conv dataframe based on Different Credit Converted Value

yn = conv[conv['Credit Converted']=='N']
yy = conv[conv['Credit Converted']=='Y']

#Filter yy dataframe based on the values available in yn dataframe

yy = yy[yy['Course Prefix'].isin(yn['Course Prefix'].values)]

#reseting the yy dataframe index 
yy = yy['Student'].reset_index(drop=True)

#Modify the Student Column name in yy dataframe
yy = pd.DataFrame({'num':yy.values})
```

```python
#concating the yn and yy dataframe column wise
rate = pd.concat([yn,yy],axis=1)
#creating a new column which calcuates the conversion rate percentage
rate['conversion_rate'] = rate['num']/(rate['Student']+rate['num']) * 100
rate = rate[['Course Prefix', 'Student','num','conversion_rate']]
rate
```

```python
rate.to_csv('conversion_rate.csv')
```

```python
#Grouping by the dataframe based on term, crediteligible, creditconverted and courseprefix column which has credit Eligible 'Y' value
term_conv = df[df['Credit Eligible']=='Y'].groupby(['Term','Credit Eligible','Credit Converted','Course Prefix']).agg({'Student':pd.Series.count})
term_conv = term_conv.reset_index()
term_conv
```

```python
#Creating a dataframe which only contains Credit Converted "N" values
ccn = term_conv[term_conv['Credit Converted']=='N']

#Filtering the columns available in ccn dataframe
ccn = ccn[['Term','Course Prefix','Student']]

#reseting the index 
ccn = ccn.set_index(['Term','Course Prefix'])
ccn
```

```python
#Creating a dataframe which only contains Credit Converted "Y" values
ccy = term_conv[term_conv['Credit Converted']=='Y']
#Filtering the columns available in ccy dataframe
ccy = ccy[['Term','Course Prefix','Student']]
#reseting the index
ccy = ccy.set_index(['Term','Course Prefix'])
ccy
```

```python
#performing outer join on the ccn and ccy dataframe
convterm = pd.merge(left=ccy,right=ccn,how='outer',left_index=True,right_index=True)

convterm['conversion_rate'] = convterm['Student_x']/(convterm['Student_x']+convterm['Student_y']) * 100
```

```python
convterm
```

```python
convterm.to_csv("termwise_conversion.csv")
```
